```python
import keras
from keras.preprocessing.image import *
```

```python
yes_dir = '/content/drive/MyDrive/Interview NexTlab/brain_tumor_dataset/yes'
no_dir = '/content/drive/MyDrive/Interview NexTlab/brain_tumor_dataset/no'
```

```python
base_data = '/content/drive/MyDrive/Interview NexTlab/brain_tumor_dataset'
```

```python
# small dataset so using image augmentation
datagen = ImageDataGenerator(rescale = 1/255,
                              rotation_range = 20,
                              horizontal_flip=True,
                              height_shift_range=0.1,
                              width_shift_range=0.1,
                              shear_range=0.1,
                              brightness_range=[0.3,1.5],
                              validation_split=0.2)
```

```python
# creating flow from dic
train = datagen.flow_from_directory(base_data,
                                    target_size=(224,224),
                                    class_mode='binary',
                                    subset='training'
                                    )


test = datagen.flow_from_directory( base_data,
                                  target_size=(224,224),
                                  class_mode='binary',
                                subset='validation'
                                  )
```

```python
from keras.models import * #(* import all the models i.e. sequential as well as functional)
from keras.layers import *
import tensorflow as tf

vgg = tf.keras.applications.VGG16

vgg16 = vgg(weights= 'imagenet', # one of None (random initialization), 'imagenet' (pre-training on ImageNet), or the path to the weights file to be loaded.
            include_top = False, #have to remove this because of the binary class classification in the given problem
            input_shape = (224,224, 3)
)

vgg.trainable = False
model = Sequential()
model.add(vgg16)
model.add(Flatten())
model.add(Dense(units = 128, activation='relu'))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(Dense(units=1, activation= 'sigmoid'))
model.compile(loss='binary_crossentropy', metrics=['accuracy'])

store = model.fit(train, epochs =30, batch_size=323,
                  validation_data=test)

early_stopping = keras.callbacks.EarlyStopping(
    patience=10,
    min_delta=0.001,
    restore_best_weights=True,
)
```
